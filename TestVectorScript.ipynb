##Subtraction

```python
def bindigits(n, bits):
    s = bin(n & int("1"*bits, 2))[2:]
    return ("{0:0>%s}" % (bits)).format(s)

X=0
F=1
while (X < 16):
  for Y in range(16):
    result=bindigits(X-Y,4)
    print(str(format(F,'#04b'))[2:] +' ' + str(format(X,'#06b'))[2:] + ' ' +  str(format(Y,'#06b')) [2:]+ ' ' +  result)
   
  X=X+1 
```

##XNOR

```python
X=0
F=3
while (X < 16):
  for Y in range(16):
    result=bindigits( ~(X^Y),4)
    print(str(format(F,'#04b'))[2:] +' ' + str(format(X,'#06b'))[2:] + ' ' +  str(format(Y,'#06b')) [2:]+ ' ' +  result)
   
  X=X+1 
```
